```python
feature_dict = {i:label for i,label in zip(
                range(4),
                  ('sepal length in cm',
                  'sepal width in cm',
                  'petal length in cm',
                  'petal width in cm', ))}

import pandas as pd

df = pd.read_csv('iris.data')
df.columns = [l for i,l in sorted(feature_dict.items())] + ['class label']
df.dropna(how="all", inplace=True) # to drop the empty line at file-end

df.head()
```

```python
## 对数据进行预处理
from sklearn.preprocessing import LabelEncoder
x = df.iloc[:,0:4].values
y = df.iloc[:,-1].values
```

```python
## 进行数据预处理
enc = LabelEncoder()
label_encoder = enc.fit(y)
y = label_encoder.transform(y) + 1
```

```python
y
```

```python
label_dict = {1: 'Setosa', 2: 'Versicolor', 3:'Virginica'}
```

```python
import numpy as np
np.set_printoptions(precision=4)
mean_vectors = []
for cl in range(1,4):
    print(x[y==cl])
    mean_vectors.append(np.mean(x[y ==cl],axis=0))
    print("Mean Vector class %s: %s\n" %(cl,mean_vectors[cl-1]))
```

```python
S_W = np.zeros((4,4))
for cl,mv in zip(range(1,4),mean_vectors):
    class_sc_mat = np.zeros((4,4)) ## scatter matrix for every class
    for row in x[y ==cl]:
        row,mv = row.reshape(4,1),mv.reshape(4,1)
        class_sc_mat += (row-mv).dot((row-mv).T)
    S_W += class_sc_mat
print("With-class Scatter Matrix;\n",S_W)
```

```python
overall_mean = np.mean(x,axis=0)
```

```python
overall_mean
```

```python
S_B = np.zeros((4,4))
S_B
```

```python
x
```

```python
for i,mean_vec in enumerate(mean_vectors):
    print(i,mean_vec)
```

```python
for i,mean_vec in enumerate(mean_vectors):
    n = x[y == i+1,:].shape[0]
    mean_vec = mean_vec.reshape(4,1)
    overall_mean = overall_mean.reshape(4,1)
    S_B += n*(mean_vec - overall_mean).dot((mean_vec - overall_mean).T)
print("between-class Scatter Matrix:\n",S_B)
```

```python
##  求解矩阵的特征值
## np.linalg.inv()：矩阵求逆
eig_vals,eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
eig_vals
```

```python
eig_vecs
```

```python
for i in range(len(eig_vals)):
    eigvec_sc = eig_vecs[:,i].reshape(4,1)
    print("\nEigenvector {}:\n{}".format(i+1,eigvec_sc.real))
    print("Eigenvalue {:}:{:.2e}".format(i+1,eig_vals[i].real))
    
```

```python
eig_pairs = [(np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs = sorted(eig_pairs,key=lambda k:k[0],reverse=True)
print("Eigenvalues in decreasing order:\n")
for i in eig_pairs:
    print(i[0])
```

```python
print("Variance explained:\n")
eigv_sum = sum(eig_vals)
for i,j in enumerate(eig_pairs):
    print("eigenvalue {}:{:.2%}".format(i+1,(j[0]/eigv_sum).real))
```

```python
## 选择前2维特征
W = np.hstack((eig_pairs[0][1].reshape(4,1),eig_pairs[1][1].reshape(4,1)))
print("Matrix W:\n",W.real)
```

```python
X_lda = x.dot(W)
X_lda
```

```python
import  matplotlib.pyplot as plt
def plot_step_lda():
    ax = plt.subplot(111)
    for label,marker,color in zip(range(1,4),('^','s','o'),('blue','red','green')):
        plt.scatter(x =X_lda[:,0].real[y ==label],y = X_lda[:,1].real[y == label],
        marker = marker,color = color,alpha= 0.5,label = label_dict[label])
    plt.xlabel("LD1")
    plt.ylabel("LD2")
    leg = plt.legend(loc="upper right",fancybox=True)
    leg.get_frame().set_alpha(0.5)
    plt.title("LDA:Iris projection onto the first 2 linear discriminants")
    ## hide  axis ticks
    plt.tick_params(axis="both",which = "both",bottom = "off",top ="off",labelbottom="on",left = "off",right ="off",labelleft ="on")
    ## remove axis spines
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines["left"].set_visible(False)
    plt.grid()
    plt.tight_layout
    plt.show()
plot_step_lda()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
sklearn_lda = LDA(n_components=2)
X_lda_sklearn = sklearn_lda.fit_transform(x,y)
```

```python
def plot_scikit_lda(X, title):

    ax = plt.subplot(111)
    for label,marker,color in zip(
        range(1,4),('^', 's', 'o'),('blue', 'red', 'green')):

        plt.scatter(x=X[:,0][y == label],
                    y=X[:,1][y == label] * -1, # flip the figure
                    marker=marker,
                    color=color,
                    alpha=0.5,
                    label=label_dict[label])

    plt.xlabel('LD1')
    plt.ylabel('LD2')

    leg = plt.legend(loc='upper right', fancybox=True)
    leg.get_frame().set_alpha(0.5)
    plt.title(title)

    # hide axis ticks
    plt.tick_params(axis="both", which="both", bottom="off", top="off",  
            labelbottom="on", left="off", right="off", labelleft="on")

    # remove axis spines
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)    

    plt.grid()
    plt.tight_layout
    plt.show()
```

```python
plot_step_lda()
plot_scikit_lda(X_lda_sklearn,title ="Default LDA via scikit-learn")
```
